# Dynamic Analysis

###Â 1. Degree Distribution:

Visualise the degree evolution of a node, let's say the first node:

```python
degrees = S.nodes_degree()
```

```python
def cluster_as_time_serie(cluster):
    times = []
    values = []
    for n in cluster:
        times += [n[0],n[1]]
        values += [n[2],n[2]]
    return times,values  
```

```python
times,values = cluster_as_time_serie(degrees[0])
plt.plot(times,values)
```

```python
times,values = cluster_as_time_serie(degrees[1])
plt.plot(times,values)
```
