## Property visualisation

We can represent clusters (sets of temporal nodes) with the same function:

```python
clusters = [[(1,2,'A'),(2,3,'B')],[(2,3,'D'),(2,3,'F')]]
S.plot(clusters = clusters)
```

We can represent a given value from a set of temporal nodes with the following method:

```python
dict_prop_to_cluster = {4:[(0,1,'A'),(3,4,'A'),(1,4,'B'),(0,1,'D')],8:[(0,3,'E'),(0,3,'F')],0:[(7,8,'C')]}
S.plot_dict_clusters(dict_prop_to_cluster,title= "Example Property")
```
